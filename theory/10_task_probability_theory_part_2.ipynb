### Немного о прошлом

В прошлый раз рассмотрели вероятность одного события. Вспомним формулу.

Пусть у нас есть событие $A$, тогда ее вероятность:

$$
P(A) = ?
$$

### А что, если событий несколько?

На самом деле для любого эксперимента может существовать несколько событий. Например, мы можем бросать монетку и у нас может быть два события - 1. выпадет решка и 2. выпадет орел.

Иногда нам может понадобиться рассмотреть совокупность событий. Существует несколько видов событий, рассмотрим их.

### 1. Несовместные события

Несовместными событиями называются те, которые не могут случиться одновременно.

Если рассмотреть бросание кубика, то у нас не могут одновременно выпасть, допустим, и 1, и 2.

Допустим, у нас есть событие $A$ и событие $B$, они несовместны. И мы задаемся вопросом, а какова вероятность того, что произойдет или событие $A$, или событие $B$? 

Для события $A$ есть вероятность $ - $ это $P(A)$. Для события $B$ $ - $ $P(B)$. И мы хотим найти $P(A \text{ или } B)$.

#### Формула вероятности несовместных событий

$$
P(A \text{ или } B) = P(A) + P(B)
$$

### **Пример 1**

<img src="https://www.astroguide.ru/files/article/41864/main-image/0305_3kub.jpg" style="width: 400px">

Рассмотрим бросание кубика. Допустим, мы хотим найти вероятность того, что после броска выпадет или 3, или 4. То есть у нас есть два события $ - $ выпадет 3 и выпадет 4.

Понятно, что они несовместны, потому что после бросания кубика у нас может выпасть только одно число, а не сразу два.

**Пробуем обозначить все, что имеем**

*Эксперимент:* бросание кубика.

*Исходы:* 1, 2, 3, 4, 5, 6

*Количество всех исходов:* 6

*Событие $A$:* выпадет 3.

*Событие $B$:* выпадет 4.

*Благоприятные исходы для события $A$:* 3

*Количество благоприятных исходов для события $A$:* 1

*Благоприятные исходы для события $B$:* 4

*Количество благоприятных исходов для события $B$:* 1

**Подсчет вероятности**

$$
P(A) = \frac{1}{6}
$$

$$
P(B) = \frac{1}{6}
$$

Тогда:

$$
P(A \text{ или } B) = P(A) + P(B) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}
$$

### 2. Противоположные события

Противоположные события $ - $ события, которые несовместны, но одно из них обязательно произойдет.

Разница между просто несовместными и противоположными это то, что несоместные могут и не произойти, а среди противоположных одно произойдет точно.

Например, если мы подбрасываем монетку, то выпадет или орел, или решка. Одно из них обязательно произойдет. А второе произойти не сможет.

#### Сумма противоположных событий

Допустим, есть событие $A$ и событие $B$, они противоположны. Тогда сумма их вероятностей равна 1:

$$
P(A) + P(B) = 1
$$

### Пример 2.

<img src="https://iron-bet.ru/upload/iblock/c93/c931e7ded72840261c5687905b72845f.jpg" style="width: 400px">

Рассмотрим бросание монетки. Мы знаем, что вероятность выпадения орла равна $\frac{1}{2}$. Чему равна вероятность выпадения решки?

Конечно, мы можем посчитать выпадение решки как отдельного события, как мы делали на прошлом занятии, но давайте попробуем воспользоваться формулой суммы противоположных событий.

**Пробуем обозначить все, что имеем**

*Событие $A$:* выпадение орла.

*Вероятность события $A$:* $\frac{1}{2}$

*Событие $B$:* выпадение решки.

**Подсчет вероятности**

У нас есть формула:

$$
P(A) + P(B) = 1
$$

Мы знаем $P(A)$. Тогда:

$$
P(B) = 1 - P(A) = 1 - \frac{1}{2} = \frac{1}{2}
$$

### Независимые события

События $A$ и $B$ называют независимыми, если одно из них не меняет вероятность другого.

<img src="https://strelasamara.ru/wp-content/uploads/2019/01/archery-clipart.png" style="width: 300px">

Например, у нас есть стрелок, который пытается попасть в мишень. Он попадает всегда с какой-то определенной вероятностью. И допустим, он хочет произвести два выстрела. Тогда можем рассмотреть два события $ - $ попасть в первом выстреле и попасть во втором выстреле. Так как после первого выстрела стрелок не станет стрелять хуже, то вероятность попадания останется той же. То есть первое событие никак не повлияло на вероятность второго, поэтому эти события независимые.

Пусть у нас есть событие $A$, которое имеет вероятность $P(A)$, и событие $B$, которое имеет вероятность $P(B)$. Мы знаем, что они независимы. И мы задаемся вопросом, а какова вероятность, что оба события произойдут? В таком случае мы будем искать вероятность $P(A \text{ и } B)$. 

#### Формула вероятности двух независимых событий

$$
P(A \text{ и } B) = P(A) * P(B)
$$

Часто бывает так, что у нас есть больше, чем два независимых события. Тогда вероятность всех этих событий все равно будет равна произведению. Например, для трех независимых событий: $P(A1 \text{ и } A2 \text{ и } A3) = P(A1) * P(A2) * P(A3)$ и так далее.
