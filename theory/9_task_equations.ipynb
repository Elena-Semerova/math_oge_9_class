Тема уравнений очень важна, и не только в мире математики. На самом деле очень часто в различных областях встречаются процессы, в которых присутствуют неизвестные компоненты, которые было бы весьма полезно получать для дальнейшего изучения и анализа.

Например, в экономике могут быть различные факторы, влияющие на успех какого-то явления, например, роста акций какой-то компании. В биологии могут возникать какие-то мутации генов, которые можно обнаружить и опознать.

Из этого мы можем выделить основную цель решения уравнения $ - $ найти что-то неизвестное.

Введем определения.

**Уравнение** $ - $ выражение со знаком равенства, в котором присутствует *неизвестная переменная*, относительно которой нужно найти *решение*.

**Неизвестная переменная** $ - $ это то, что мы ищем в ходе решения уравнения. Обознается буквами.

**Решение** $ - $ это числовые значения, которые может принимать неизвестная переменная, чтобы при подстановке этих чисел вместо нее в уравнение получалось тождество.

### Пример 1

У нас есть такое *уравнение*:

$$
x + 7 = 18
$$

*Неизвестная переменная* в нем обозначена $x$.

*Решением* будет являться число 11.

То есть при $x = 11$ при подстановке в уравнение мы получим:

$$
11 + 7 = 18
$$

Это и правда является *тождеством*.

### Виды уравнений

Существует несколько видов уравнений, сегодня мы рассмотрим следующие:

1. Линейные уравнения

2. Квадратные уравнения

3. Дробно-рациональные уравнения

Также отдельно от этого мы рассмотрим простые системы уравнений.

### 1 Линейные уравнения

Такие уравнения имеют вид $ax = b$, где на месте $a$, $b - $ стоят какие-то числа. 

Конечно, это упрощенный вид таких уравнений. Оно может усложняться при помощи разных конструкций (например, выделение каких-нибудь скобок) и при помощи "лишних" слагаемых со степенями, которые должны сократиться (если не сокращаются, то это другой вид уравнения).

**Алгоритм решения таких уравнений:**

1. Раскрыть скобки, если они имеются.
2. Перенести числа в одну часть уравнения, неизвестные переменные – в другую. Чаще всего числа переносят вправо, а буквы влево. Не забываем при этом менять знаки на противоположные при переносе через знак равенства.
3. Привести подобные слагаемые, то есть сложить все, что складывается, вычесть все, что вычитается и т.д.
4. Увидеть, что получили уравнение вида aх=b.
5. Привести это уравнение к виду: $х = \frac{b}{а}$ (если $a = 0$ корней нет) и посчитать.

### Пример 2

Имеем уравнение:

$$
5 - 2x = 11 - 7(x + 2).
$$

Выполним решение по алгоритму:

1. $5 - 2x = 11 - 7x - 14$

2. $-2x + 7x = 11 - 14 - 5$

3. $5x = -8$

4. Увидели. У нас $a = 5$, $b = -8$

4. $x = \frac{-8}{5} = -1.6$

Готово!

### 2 Квадратные уравнения

Такие уравнения имеют вид $ax^2 + bx + c = 0$, где $a, b, c - $ какие-то числа.

Для решения таких уравнений вводится такое понятие, как дискриминант. Он обозначается буквой $D$ и считается по следующей формуле:

$$
D = b^2 - 4 ac
$$

Если $D > 0$, то решений у нас два и они вычисляются по следующим формулам:

$$
x_1 = \frac{-b + \sqrt{D}}{2a}; x_2 = \frac{-b - \sqrt{D}}{2a}
$$

Если $D = 0$, то решение у нас одно, оно вычисляется по формуле:

$$
x = \frac{-b}{2a}
$$

Если $D < 0$, то решений нет.

**Алгоритм решения таких уравнений:**

1. Раскрыть скобки, если они имеются.
2. Перенести все в левую часть уравнения, чтобы справа остался только 0.
3. Привести подобные слагаемые.
4. Посчитать дискриминант по формуле.
5. Определить, какой дискриминант: $D > 0$, $D = 0$ или $D < 0$.
6. В зависимости от дискриминанта посчитать решения или написать, что их нет.

### Пример 4

Имеем уравнение:

$$
(−5x + 3)(−x + 6) = 0
$$

Выполним решение по алгоритму:

1. $5x^2 - 30x - 3x + 18 = 0$

2. Готово

3. $5x^2 - 33x + 18 = 0$

4. $D = b^2 - 4ac = (-33)^2 - 4 *  5 * 18 = 1089 - 360 = 729$

5. $D = 729 > 0$. Будет два корня

6. Считаем:

$x_1 = \frac{33 + \sqrt{729}}{10} = \frac{33 + 27}{10} = \frac{60}{10} = 6$

$x_2 = \frac{33 - 27}{10} = \frac{6}{10} = 0.6$

### 3 Дробно-рациональные уравнения

Эти уравнение не имеют какого-то общего описания, но их всегда можно узнать, так как в них обязательно присутствуют дроби.

**Алгоритм решения таких уравнений:**

1. Умножить обе части уравнения на общий знаменатель дробей. Если в знаменателях есть выражения с неизвестной переменной, то дополнительно записать условие на неравенство нулю.
2. Раскрыть скобки, если они имеются.
3. Перенести слагаемые с неизвестными переменными в одну сторону, числа - в другую.
4. Привести подобные слагаемые.
5. Решить получившееся уравнение по одному из предыдущих алгоритмов.

### Пример 6

Имеем уравнение:

$$
\frac{9}{x - 2} = \frac{9}{2}
$$

Выполним решение по алгоритму:

1. Общий знаменатель: $2(x - 2)$. Получим: $9 * 2 = 9 * (x - 2)$. В знаменателе есть $x - 2$. Условие на неравенство 0: $x - 2 \not= 0$ или $x \not= 2$

2. $9 * 2 = 9x - 18$

3. $18 + 18 = 9x$

4. $36 = 9x$

5. $x = \frac{36}{9} = 4$

### 4 Системы уравнений

До этого мы рассматривали по одному уравнению, но иногда могут встречаться задания, где есть несколько уравнений. На ОГЭ могут встретиться с двумя уравнениями.

Так как увеличивается количество уравнений, то увеличивается и количество неизвестных переменных. Иы хотим найти решения, которые будут выполняться одновременно для обеих переменных.

Для этого задания будем пользоваться методом подстановки для решения системы уравнений.

**Метод подстановки:**

1. Выразить из одного любого уравнения одну из переменных. 

2. Подставить значение выраженной переменной во второе уравнение.

3. Решить получившееся уравнение относительно одной неизвестной переменной.

4. Подставить получившееся решение в выражение с другой переменной и посчитать.

### Пример 8

Имеем систему уравнений:

$$
\begin{cases}
    4x - 2y = 2,\\
    2x + y = 5.\\
\end{cases}
$$

Решим при помощи метода подстановки:

1. Выразим из первого уравнения переменную $y$ :

$-2y = 2 - 4x$

$y = \frac{2 - 4x}{-2} = -1 + 2x$

2. Подставим полученное выражение во второе уравнение:

$2x + (-1 + 2x) = 5$

3. Решим:

$2x - 1 + 2x = 5$

$4x = 6$

$x = \frac{6}{4} = 1.5$

4. Подставим получившееся значение в выражение с $y$:

$y = -1 + 2(1.5) = -1 + 3 = 2$

Получили решение:

$$
\begin{cases}
    x = 1.5,\\
    y = 2.\\
\end{cases}
$$
