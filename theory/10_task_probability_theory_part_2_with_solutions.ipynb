### Немного о прошлом

В прошлый раз рассмотрели вероятность одного события. Вспомним формулу.

Пусть у нас есть событие $A$, тогда ее вероятность:

$$
P(A) = \frac{\text{количество благоприятных исходов}}{\text{количество всех исходов}}
$$

### А что, если событий несколько?

На самом деле для любого эксперимента может существовать несколько событий. Например, мы можем бросать монетку и у нас может быть два события - 1. выпадет решка и 2. выпадет орел.

Иногда нам может понадобиться рассмотреть совокупность событий. Существует несколько видов событий, рассмотрим их.

### 1. Несовместные события

Несовместными событиями называются те, которые не могут случиться одновременно.

Если рассмотреть бросание кубика, то у нас не могут одновременно выпасть, допустим, и 1, и 2.

Допустим, у нас есть событие $A$ и событие $B$, они несовместны. И мы задаемся вопросом, а какова вероятность того, что произойдет или событие $A$, или событие $B$? 

Для события $A$ есть вероятность $ - $ это $P(A)$. Для события $B$ $ - $ $P(B)$. И мы хотим найти $P(A \text{ или } B)$.

#### Формула вероятности несовместных событий

$$
P(A \text{ или } B) = P(A) + P(B)
$$

### **Пример 1**

<img src="https://www.astroguide.ru/files/article/41864/main-image/0305_3kub.jpg" style="width: 400px">

Рассмотрим бросание кубика. Допустим, мы хотим найти вероятность того, что после броска выпадет или 3, или 4. То есть у нас есть два события $ - $ выпадет 3 и выпадет 4.

Понятно, что они несовместны, потому что после бросания кубика у нас может выпасть только одно число, а не сразу два.

**Пробуем обозначить все, что имеем**

*Эксперимент:* бросание кубика.

*Исходы:* 1, 2, 3, 4, 5, 6

*Количество всех исходов:* 6

*Событие $A$:* выпадет 3.

*Событие $B$:* выпадет 4.

*Благоприятные исходы для события $A$:* 3

*Количество благоприятных исходов для события $A$:* 1

*Благоприятные исходы для события $B$:* 4

*Количество благоприятных исходов для события $B$:* 1

**Подсчет вероятности**

$$
P(A) = \frac{1}{6}
$$

$$
P(B) = \frac{1}{6}
$$

Тогда:

$$
P(A \text{ или } B) = P(A) + P(B) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}
$$

### 2. Противоположные события

Противоположные события $ - $ события, которые несовместны, но одно из них обязательно произойдет.

Разница между просто несовместными и противоположными это то, что несоместные могут и не произойти, а среди противоположных одно произойдет точно.

Например, если мы подбрасываем монетку, то выпадет или орел, или решка. Одно из них обязательно произойдет. А второе произойти не сможет.

#### Сумма противоположных событий

Допустим, есть событие $A$ и событие $B$, они противоположны. Тогда сумма их вероятностей равна 1:

$$
P(A) + P(B) = 1
$$

### Пример 2.

<img src="https://iron-bet.ru/upload/iblock/c93/c931e7ded72840261c5687905b72845f.jpg" style="width: 400px">

Рассмотрим бросание монетки. Мы знаем, что вероятность выпадения орла равна $\frac{1}{2}$. Чему равна вероятность выпадения решки?

Конечно, мы можем посчитать выпадение решки как отдельного события, как мы делали на прошлом занятии, но давайте попробуем воспользоваться формулой суммы противоположных событий.

**Пробуем обозначить все, что имеем**

*Событие $A$:* выпадение орла.

*Вероятность события $A$:* $\frac{1}{2}$

*Событие $B$:* выпадение решки.

**Подсчет вероятности**

У нас есть формула:

$$
P(A) + P(B) = 1
$$

Мы знаем $P(A)$. Тогда:

$$
P(B) = 1 - P(A) = 1 - \frac{1}{2} = \frac{1}{2}
$$

### Независимые события

События $A$ и $B$ называют независимыми, если одно из них не меняет вероятность другого.

<img src="https://strelasamara.ru/wp-content/uploads/2019/01/archery-clipart.png" style="width: 300px">

Например, у нас есть стрелок, который пытается попасть в мишень. Он попадает всегда с какой-то определенной вероятностью. И допустим, он хочет произвести два выстрела. Тогда можем рассмотреть два события $ - $ попасть в первом выстреле и попасть во втором выстреле. Так как после первого выстрела стрелок не станет стрелять хуже, то вероятность попадания останется той же. То есть первое событие никак не повлияло на вероятность второго, поэтому эти события независимые.

Пусть у нас есть событие $A$, которое имеет вероятность $P(A)$, и событие $B$, которое имеет вероятность $P(B)$. Мы знаем, что они независимы. И мы задаемся вопросом, а какова вероятность, что оба события произойдут? В таком случае мы будем искать вероятность $P(A \text{ и } B)$. 

#### Формула вероятности двух независимых событий

$$
P(A \text{ и } B) = P(A) * P(B)
$$

Часто бывает так, что у нас есть больше, чем два независимых события. Тогда вероятность всех этих событий все равно будет равна произведению. Например, для трех независимых событий: $P(A1 \text{ и } A2 \text{ и } A3) = P(A1) * P(A2) * P(A3)$ и так далее.

##### 1. В магазине канцтоваров продаётся 100 ручек, из них 37 – красные, 8 – зелёные, 17 – фиолетовые, ещё есть синие и чёрные, их поровну. Найдите вероятность того, что Алиса наугад вытащит красную или чёрную ручку.

**Имеем:**

*Эксперимент:* покупка ручек.

*Исходы:* ручки.

*Количество исходов:* 100

*Событие $A$:* вытащить красную ручку.

*Благоприятные исходы для события $A$:* красные ручки.

*Количество благоприятных исходов для события $A$:* 37

*Событие $B$:* вытащить черную ручку.

*Благоприятные исходы для события $B$:* черные ручки.

*Количество благоприятных исходов для события $B$:* $\frac{100 - (37 + 8 + 17)}{2} = \frac{38}{2} = 19$ 

**Какие у нас события $A$ и $B$?** 

Несовместные события

**Используем формулу:**

$$
P(A \text{ или } B) = P(A) + P(B)
$$

**Посчитаем:**

1. $P(A) = \frac{37}{100}$ 

2. $P(B) = \frac{19}{100}$ 

3. Наша формула: $P(A \text{ или } B) = P(A) + P(B) = \frac{37}{100} + \frac{19}{100} = \frac{56}{100} = 0.56$

```python
# тип событий
type_of_events = 'несовместные'

# вероятность события A
proba_A = 37 / 100

# вероятность события B
proba_B = 19 / 100

if type_of_events == 'несовместные':
    # вероятность, что произойдет или A, или B
    proba_A_or_B = proba_A + proba_B
    
    print('Вероятность события A или B =', proba_A_or_B)

if type_of_events == 'противоположные':
    # вероятность события B
    proba_A = 1 - proba_B
    
    print('Вероятность A =', proba_A)
    
if type_of_events == 'независимые':
    # вероятность, что произойдет A и B
    proba_A_and_B = proba_A * proba_B
    
    print('Вероятность A и B =', proba_A_and_B)
```

##### 5. Фирма «Вспышка» изготавливает фонарики. Вероятность того, что случайно выбранный фонарик из партии бракованный, равна 0,02. Какова вероятность того, что два случайно выбранных из одной партии фонарика окажутся небракованными?

**Имеем:**

*Эксперимент:* выбор фонарика.

*Исходы:* фонарики.

*Событие $A$:* вытащить первый небракованный фонарик.

*Благоприятные исходы для события $A$:* небракованные фонарики.

*Событие $B$:* вытащить второй небракованный фонарик.

*Благоприятные исходы для события $B$:* небракованные фонарики.

**Какие у нас события $A$ и $B$?** 

Независимые события

**Используем формулу:**

$$
P(A \text{ и } B) = P(A) * P(B)
$$

**Посчитаем:**

Заметим, что мы каждый раз вытягиваем небракованный фонарик. Так как события независимые, то у нас для каждого вытягивания будет одна и та же вероятность вытащить небракованный фонарик. Поэтому $P(A) = P(B)$.

Но нам надо найти $P(A)$.

В условии дано: *Вероятность того, что случайно выбранный фонарик из партии бракованный, равна 0,02*. Описанное событие является противоположным для события $A$, так как мы не можем одновременно вытащить бракованный и небракованный фонарик, сможем вытащить только что-то одно из этого. Обозначим его как событие $A1$.

Тогда пользуемся формулой: 

$$
P(A) + P(A1) = 1
$$

Отсюда: 

1. $P(A) = 1 - 0.02 = 0.98$ 

2. $P(B) = 0.98$ 

3. Наша формула: $P(A \text{ и } B) = P(A) * P(B) = 0.98 * 0.98 = 0.9604$

```python
# тип событий
type_of_events = 'независимые'

# вероятность события A
proba_A = 1 - 0.02

# вероятность события B
proba_B = proba_A

if type_of_events == 'несовместные':
    # вероятность, что произойдет или A, или B
    proba_A_or_B = proba_A + proba_B
    
    print('Вероятность события A или B =', proba_A_or_B)

if type_of_events == 'противоположные':
    # вероятность события B
    proba_A = 1 - proba_B
    
    print('Вероятность A =', proba_A)
    
if type_of_events == 'независимые':
    # вероятность, что произойдет A и B
    proba_A_and_B = proba_A * proba_B
    
    print('Вероятность A и B =', proba_A_and_B)
```

*Событие $B$:* вытащить ручку, которая пишет плохо.

Противоположные события

$$

##### 7. Стрелок 4 раза стреляет по мишеням. Вероятность попадания в мишень при одном выстреле равна 0,5. Найдите вероятность того, что стрелок первые 3 раза попал в мишени, а последний раз промахнулся.

**Имеем:**

*Эксперимент:* стрельба в мишень.

Независимые события

**Используем формулу:**

$$
P(A \text{ и } B) = P(A) * P(B)
$$

**Посчитаем:**

Заметим, что событие $A - $ это на самом деле одновременно три независимых события. То есть у нас происходит три выстрела подряд, и каждый из них не зависит от предыдущего. А мы знаем, что в независимости от количества событий, вероятность нескольких событий равна произведению вероятностей каждого события. 

Найдем вероятность попадания при одном выстреле: $1 - 0.5 = 0.5$

Тогда если $A1 - $ вероятность попадания в первом выстреле, $A2 - $ вероятность попадания во втором выстреле, $A3 - $ вероятность попадания в третьем выстреле, то:

1. $P(A) = P(A1) * P(A2) * P(A3) = 0.5 * 0.5 * 0.5 = (0.5)^3 = 0.125$ 

2. $P(B) = 0.5$ (по условию) 

3. Наша формула: $P(A \text{ и } B) = P(A) * P(B) = 0.125 * 0.5 = 0.0625$

```python
# тип событий
type_of_events = 'независимые'

# вероятность события A
proba_A = 0.5**3

# вероятность события B
proba_B = 0.5

if type_of_events == 'несовместные':
    # вероятность, что произойдет или A, или B
    proba_A_or_B = proba_A + proba_B
    
    print('Вероятность события A или B =', proba_A_or_B)

if type_of_events == 'противоположные':
    # вероятность события B
    proba_A = 1 - proba_B
    
    print('Вероятность A =', proba_A)
    
if type_of_events == 'независимые':
    # вероятность, что произойдет A и B
    proba_A_and_B = proba_A * proba_B
    
    print('Вероятность A и B =', proba_A_and_B)
```

##### 12. В группе из 20 российских туристов несколько человек владеют иностранными языками. Из них пятеро говорят только по-английски, трое только по-французски, двое по-французски и по-английски. Какова вероятность того, что случайно выбранный турист говорит по-французски?

**Имеем:**

*Эксперимент:* выбор туриста.

*Исходы:* туристы.

*Количество исходов:* 20

*Событие $A$:* выберем туриста, который говорит по-французски.

*Благоприятные исходы для события $A$:* турист говорит только по-французски, турист говорит по-французски и по-английски.

*Количество благоприятных исходов для события $A$:* 5

**Используем формулу:**

$$
P(A) = \frac{\text{количество благоприятных исходов}}{\text{количество всех исходов}} = \frac{5}{20} = 0.25
$$

```python
# количество благоприятных исходов:
number_of_desired_outcomes = 5

# количество всех исходов
number_of_all_outcomes = 20

# считаем вероятность:
probability = number_of_desired_outcomes / number_of_all_outcomes

print('Вероятность события A =', probability)
```
