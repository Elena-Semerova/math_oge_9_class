### **Что такое вероятность?**

**Вероятность** - это то, на сколько процентов (или на какую долю) возможно наступление какого-то события. 

Сразу запомним, что **вероятность лежит в промежутке от 0 до 1**. То есть у нас не может быть вероятность меньше 0, и не может быть больше 1. 

Когда будем решать задачки, об этом надо будет помнить, потому что если будет получаться ответ < 0 или > 1, то мы что-то решили неправильно.

Обычно у нас есть какой-то **эксперимент**, который может давать разные итоги. Эти итоги называются **исходами**. 

**Событие** - это то, что может случиться в ходе эксперимента. Для события есть благоприятные исходы (обычно только часть из всех возможных исходов).

Вероятность события считается, как отношение числа благоприятных исходов к общему числу исходов.

### **Обозначения**

Событие называют какой-то буквой, например, $A$.

Вероятность события $A$ обозначают так: $P(A)$.

Тогда вероятность считается следующим образом:

$$
P(A) = \frac{\text{количество благоприятных исходов}}{\text{количество всех исходов}}
$$

### **Пример 1**

<img src="https://img3.badfon.ru/original/2560x1600/a/2c/moneta-vybor-fon.jpg" style="width: 400px">

Посмотрим на подбрасывание монетки. У нас это будет являться экспериментом.

Как мы знаем, у нас может выпасть либо орел, либо решка. Значит, в ходе этого эксперимента мы имеем два возможных исхода.

Допустим, Настя загадала, что выпадет орел. Тогда интересующее нас событие - это выпадение орла. Мы задаемся вопросом: какова вероятность, что выпадет орел?

Понятно, что благоприятный исход для этого события - выпадение орла, тогда Настя выиграет.

**Подведем небольшие итоги, что мы имеем:**

*Эксперимент:* подбрасывание монетки.

*Исходы:* выпадение орла, выпадение решки.

*Количество всех исходов:* 2

*Событие $A$:* выпадет орел.

*Благоприятный исход для события $A$:* выпадение орла.

*Количество благоприятных исходов:* 1

**Подсчет вероятности**

У нас есть формула:

$$
P(A) = \frac{\text{количество благоприятных исходов}}{\text{количество всех исходов}}
$$

Мы уже обозначили количество благоприятных исходов и общее количество исходов. Подставим эти числа и получим вероятность:

$$
P(A) = \frac{1}{2} = 0.5
$$

Отсюда получаем, что вероятность того, что действительно выпадет орел - $\frac{1}{2}$ (или 50%).

Допустим, Захар любит четные числа (те, что делятся на 2). И он хочет узнать, а какова вероятность того, что если мы кинем кубик, то выпадет какое-нибудь четное число?

**Пробуем обозначить все, что имеем**

*Эксперимент:* подбрасывание кости.

У нас есть формула:

$$
P(A) = \frac{\text{количество благоприятных исходов}}{\text{количество всех исходов}}
$$

Получаем:

$$
P(A) = \frac{3}{6} = \frac{1}{2}
$$

##### 1. На экзамене 25 билетов, Сергей не выучил 3 из них. Найдите вероятность того, что ему попадётся выученный билет.

*Эксперимент:* вытягивание билетов на экзамене.

*Исходы:* экзаменационные билеты.

*Количество всех исходов:* 25

*Событие $A$:* выпадение выученного билета.

*Благоприятные исходы для события $A$:* выученные билеты.

*Количество благоприятных исходов:* 25 - 3 = 22

$$
P(A) = \frac{\text{количество благоприятных исходов}}{\text{количество всех исходов}} = \frac{22}{25}
$$

```python
# количество благоприятных исходов:
number_of_desired_outcomes = 22

# количество всех исходов
number_of_all_outcomes = 25

# считаем вероятность:
probability = number_of_desired_outcomes / number_of_all_outcomes

print('Вероятность события A =', probability)
```
